#### b(iii): Selecting the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
train_1 = pd.DataFrame(df.loc[df['class'] == 1]).iloc[0:140]
train_0 = pd.DataFrame(df.loc[df['class'] == 0]).iloc[0:70]
train = pd.concat([train_1, train_0], ignore_index = True)
print(train)
```

```python
test_1 = pd.DataFrame(df.loc[df['class'] == 1]).iloc[140:]
test_0 = pd.DataFrame(df.loc[df['class'] == 0]).iloc[70:]
test = pd.concat([test_1, test_0], ignore_index = True)
print(test)
```

```python
X_train = np.array(train.drop(['class'], 1))
y_train = np.array(train['class'])
X_test = np.array(test.drop(['class'], 1))
y_test = np.array(test['class'])
best_k_learns_test = [1,6,1,1,16,31,26,26,31]
```

#### c (i): k-nearest neighbors with Euclidean metric

```python
from sklearn import neighbors

k_val = np.arange(208, 0, -3)
train_scores = []
test_scores = []
for k in k_val: 
    neighbor = neighbors.KNeighborsClassifier(n_neighbors=k ,metric = 'euclidean')
    neighbor.fit(X_train,y_train)
    y_pred = neighbor.predict(X_test)
    train_score = neighbor.score(X_train,y_train)
    test_score = neighbor.score(X_test,y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)

train_score_np = np.array(train_scores)
test_score_np = np.array(test_scores)
#print(train_score_np)
train_error = 1 - train_score_np
#print(train_error)
test_error = 1 - test_score_np
print(test_error)
plt.figure(figsize=(27,10))
plt.plot(train_error, color='red',label='Training Error')
plt.plot(test_error, color='blue',label='Testing Error')

plt.xticks(range(len(k_val)), k_val)
plt.legend(labels = ['Train Error', 'Test Error'])
plt.xlabel('K values')
plt.ylabel('Error')
plt.title('Performace Under Varying K Values')

# best k = 4 (lowest error = 0.06)
```

#### c (ii): KNN Metrics: confusion matrix, true positive rate, true negative rate, precision,and F1-score when k = k* 

```python
from sklearn.metrics import f1_score, classification_report, accuracy_score, confusion_matrix
best_neighbor = neighbors.KNeighborsClassifier(n_neighbors = 4) #using k = 4 as inferred from above plot
best_neighbor.fit(X_train, y_train)
y_pred_best = best_neighbor.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))

CM_best = confusion_matrix(y_test, y_pred_best)

TN_best = CM_best[0][0]
FN_best = CM_best[1][0]
TP_best = CM_best[1][1]
FP_best = CM_best[0][1]

print("True positive rate:", TP_best/100)
print("True negative rate:", TN_best/100)
```

#### c (iii): Learning Curve

```python
# C (iii)
best_k_learn_test = []
N = np.arange(10, 201, 10)
for n in N:
    train_learn1 = pd.DataFrame(train.loc[train['class'] == 0]).iloc[0:int(0.3*n)]
    train_learn2 = pd.DataFrame(train.loc[train['class'] == 1]).iloc[0:int(0.7*n)]
    train_learn = pd.concat([train_learn1, train_learn2], ignore_index = True)
    #print(train_learn)
                    
    test_learn1 = pd.DataFrame(test.loc[test['class'] == 0]).iloc[int(0.3*n):]
    test_learn2 = pd.DataFrame(test.loc[test['class'] == 1]).iloc[int(0.7*n):]
    test_learn = pd.concat([test_learn1, test_learn2], ignore_index = True)
    #print(test_learn)

    X_train_learn = np.array(train_learn.drop(['class'], 1))
    y_train_learn = np.array(train_learn['class'])
    

    X_test_learn = np.array(test_learn.drop(['class'], 1))
    y_test_learn = np.array(test_learn['class'])
    #print(test_learn1)
    
    k = np.arange(1, n+1, 5)
    
    test_scores_learn = []
    test_errors_learn = np.empty(len(k))
    best_k_learn = []
    for i in k:
        learn_neighbor = neighbors.KNeighborsClassifier(n_neighbors = i)
        learn_neighbor.fit(X_train_learn, y_train_learn)
        #y_pred_learn = learn_neighbor.predict(X_test_learn)
        train_score_learn = learn_neighbor.score(X_train_learn,y_train_learn)
        test_score_learn = learn_neighbor.score(X_train_learn,y_train_learn)
        test_scores_learn.append(test_score_learn)
        test_score_learn_np = np.array(test_scores_learn)
        test_error_learn = 1 - test_score_learn_np
    
    best_k_learn_test.append(k[np.where(test_error_learn == np.amin(test_error_learn))])
```

#### using the k values with min test error

```python
k_learn_best = np.array(best_k_learns_test)
N_best = np.arange(10, 201, 10)

train_scores_best = []
test_scores_best = []
for num in N_best:
    train_learn_best1 = pd.DataFrame(train.loc[train['class'] == 0]).iloc[0:int(0.3*num)]
    train_learn_best2 = pd.DataFrame(train.loc[train['class'] == 1]).iloc[0:int(0.7*num)]
    train_learn_best = pd.concat([train_learn_best1, train_learn_best2], ignore_index = True)
    #print(train_learn_best)
                    
    test_learn_best1 = pd.DataFrame(test.loc[test['class'] == 0]).iloc[int(0.3*num):]
    test_learn_best2 = pd.DataFrame(test.loc[test['class'] == 1]).iloc[int(0.7*num):]
    test_learn_best = pd.concat([test_learn_best1, test_learn_best2], ignore_index = True)
    #print(test_learn_best)

    X_train_learn_best = np.array(train_learn_best.drop(['class'], 1))
    y_train_learn_best = np.array(train_learn_best['class'])
    #print(X_train_learn_best)

    X_test_learn_best = np.array(test_learn_best.drop(['class'], 1))
    y_test_learn_best = np.array(test_learn_best['class'])
    
    best_neighbor = neighbors.KNeighborsClassifier(n_neighbors = 1)
    best_neighbor.fit(X_train_learn_best, y_train_learn_best)
    y_pred_learn_best = best_neighbor.predict(X_train)
    train_score_best = best_neighbor.score(X_train,y_train)
    y_pred_test_best = best_neighbor.predict(X_test)
    test_score_best = best_neighbor.score(X_test,y_test)
    train_scores_best.append(train_score_best)
    test_scores_best.append(test_score_best)

train_score_best_np = np.array(train_scores_best)
test_score_best_np = np.array(test_scores_best)
#print(train_score_np)
train_error_best = 1 - train_score_best_np
#print(train_error)
test_error_best = 1 - test_score_best_np
#print(train_error_best)
plt.figure(figsize=(27,10))
#plt.plot(train_error_best, color='red',label='Training Error')
plt.plot(test_error_best, color='blue',label='Testing Error')

plt.xticks(range(len(N_best)), N_best)
plt.legend(labels = ['Test Error'])
plt.xlabel('K values')
plt.ylabel('Error')
plt.title('Performace Under Varying K Values')
```
